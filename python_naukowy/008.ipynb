```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

sp.init_printing()
```

```python
x = sp.Function('x')
t = sp.symbols('t', real=True)
w0 = sp.symbols('w0', real=True, positive=True)
beta = sp.symbols('beta', real=True, positive=True)

eq = sp.Eq(x(t).diff(t, 2) + 1/beta * x(t).diff(t) + w0**2 * x(t), 0)
eq
```

```python
eq_sol = sp.dsolve(eq, x(t), ics={x(0): 1, x(t).diff(t).subs(t, 0): 0})
eq_sol
```

```python
betas = [0.05, 1, 5]
t_arr = np.linspace(0, 15, 1000)

for beta_val in betas:
    sol = sp.lambdify(t, eq_sol.subs({w0: 5, beta: beta_val}).rhs)
    plt.plot(t_arr, sol(t_arr), label=f'beta={beta_val}')

plt.legend()
```

```python
x = sp.Function('x')
t = sp.symbols('t', real=True)
w0 = sp.symbols('w0', real=True, positive=True)
beta = sp.symbols('beta', real=True, positive=True)
a0 = sp.symbols('alpha_0', real=True, positive=True)
omega = sp.symbols('omega', real=True, positive=True)

eq = sp.Eq(x(t).diff(t, 2) + 1/beta * x(t).diff(t) + w0**2 * x(t), a0 * sp.sin(omega * t))
eq
```

```python
eq_sol = sp.dsolve(eq, x(t), ics={x(0): 0, x(t).diff(t).subs(t, 0): 0})
eq_sol
```

```python
t_arr = np.linspace(0, 10, 1000)
omega_arr = [0.5, 1, 5, 10]

for omega_val in omega_arr:
    sol = sp.lambdify(t, eq_sol.subs({w0: 5, beta: 1, a0: 1, omega: omega_val}).rhs)
    plt.plot(t_arr, sol(t_arr), label=f'omega={omega_val}')

plt.legend()
```
